```python
#imports

#we import from the nltk library the package that let us tokenize the text easily.
from nltk.tokenize import wordpunct_tokenize
#we import from the nltk library the package FreqDist that let us make the frequency distribution.
from nltk.probability import FreqDist
#from the package pathlib, we import Path to be able to access a directory
from pathlib import Path
#we import os and glob, two packages that will let us iterate over files on a local directory
import os, glob
#we import matplotlib.plyplot to be able to plot the FreqDist
import matplotlib.pyplot as plt
```

```python
#creating a variable with the path of the folder where the txt files are.
folder_path = '/content/corpus'
#defining a function to open, read and tokenize each txt file.
def corpus_analyzer(path):
  for filename in glob.glob(os.path.join(folder_path, '*.txt')):
    with open(filename, 'r') as f:
      #read the file
      text = f.read()
      #tokenize the text
      text_tokenized = wordpunct_tokenize(text)
      #normalize the text (remove punctuation)
      text_normalized = [w.lower() for w in text_tokenized if w.isalpha()]
      #take the length of the words
      word_lengths = [len(word) for word in text_normalized]
      #create a frequency distribution of the length of the words
      word_length_freq_dist = FreqDist(word_lengths)
      #create the plot of the frequency distribution
      ax = word_length_freq_dist.plot(show=False, title=filename[16:])
      #set x and y axis names
      ax.set_xlabel("length of the token")
      ax.set_ylabel("number of appearances")
      #show plot
      plt.show()

corpus_analyzer(folder_path)
```
